```python

import numpy as np
def compute_ov_matrices_gqa(model, layer_idx):
    """
    Compute OV matrices for a specific transformer layer with GQA
    
    Args:
        model: The transformer model (e.g., from HuggingFace)
        layer_idx: The index of the transformer layer to analyze
        
    Returns:
        Dictionary containing OV matrices for each query head
    """
    # Get the attention layer (implementation depends on specific model architecture)
    # This example assumes a HuggingFace model structure
    attention = model.model.layers[layer_idx].self_attn
    
    # Get the number of query heads and KV heads
    n_query_heads = 12
    n_kv_heads = 2
    
    # Get the query, key, value, and output projection weights
    # Shapes may vary depending on the specific model implementation
    q_proj = attention.q_proj.weight  # Shape: [hidden_size, hidden_size]
    k_proj = attention.k_proj.weight  # Shape: [hidden_size, hidden_size // (n_query_heads // n_kv_heads)]
    v_proj = attention.v_proj.weight  # Shape: [hidden_size, hidden_size // (n_query_heads // n_kv_heads)]
    o_proj = attention.o_proj.weight  # Shape: [hidden_size, hidden_size]
    
    hidden_size = q_proj.shape[0]
    head_dim = hidden_size // n_query_heads
    
    # Reshape the projection matrices to extract per-head weights
    # For query: each query head has its own weight matrix
    q_heads = q_proj.view(hidden_size, n_query_heads, head_dim).permute(1, 0, 2)
    # For key-value: weights are shared among query heads within the same group
    v_heads = v_proj.view(hidden_size, n_kv_heads, head_dim).permute(1, 0, 2)
    o_heads = o_proj.view(n_query_heads, head_dim, hidden_size)
    
    # Compute which KV head serves each query head
    # In GQA, query heads are evenly distributed across KV heads
    queries_per_kv = n_query_heads // n_kv_heads
    kv_head_mapping = np.repeat(np.arange(n_kv_heads), queries_per_kv)
    
    # Compute OV matrices for each query head
    ov_matrices = {}
    for query_idx in range(n_query_heads):
        kv_idx = kv_head_mapping[query_idx]
        
        # Extract the value matrix for this KV group
        v_matrix = v_heads[kv_idx]  # Shape: [hidden_size, head_dim]
        
        # Extract the output matrix for this query head
        o_matrix = o_heads[query_idx]  # Shape: [head_dim, hidden_size]
        
        # Compute the OV matrix for this query head
        ov_matrix = torch.matmul(v_matrix, o_matrix)  # Shape: [hidden_size, hidden_size]
        
        ov_matrices[query_idx] = ov_matrix
    
    return ov_matrices
ov_matrices=compute_ov_matrices_gqa(model,0)
token_to_find_contribution=2
sum=0
for i in range(15,27):
    ov_matrices=compute_ov_matrices_gqa(model,i)
    print(i)
    for j,matrix in ov_matrices.items():
        
        
        k=find_attention_length(token_to_find_contribution)
        for s in range(k):
         p,t= get_ith_token_contribution(s,i,1)
         l=get_attention_weights(s,j,i,token_to_find_contribution)
         new=p*s
         indp=t@matrix
         sum=sum+(new*indp)
print(sum)            
         
         
            
    
        
```
